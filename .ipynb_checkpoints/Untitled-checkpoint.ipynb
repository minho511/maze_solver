### 3.1. 함수 transform()
- [참고 링크](https://pyimagesearch.com/2014/08/25/4-point-opencv-getperspective-transform-example/)


```python

def order_points(pts):

    rect = np.zeros((4, 2), dtype = "float32")

    s= np.sum(pts, axis = 1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    
    d = np.diff(pts, axis = 1)
    rect[1] = pts[np.argmin(d)]
    rect[3] = pts[np.argmax(d)]
    
    return rect
    
def transform(img, pts):
    
    rect = order_points(pts)
    (tl, tr, br, bl) = rect	
    
    width_1 = np.sqrt(((br[0] - bl[0]) ** 2) + ((br[1] - bl[1]) ** 2))	
    width_2 = np.sqrt(((tr[0] - tl[0]) ** 2) + ((tr[1] - tl[1]) ** 2))
    maxWidth = max(int(width_1), int(width_2))
    
    height_1 = np.sqrt(((br[0] - tr[0]) ** 2) + ((br[1] - tr[1]) ** 2))	
    height_2 = np.sqrt(((bl[0] - tl[0]) ** 2) + ((tl[1] - bl[1]) ** 2))
    maxHeight = max(int(height_1), int(height_2))
    
    dst = np.array([
             [0, 0],
             [maxWidth - 1, 0],
             [maxWidth - 1, maxHeight - 1],
             [0, maxHeight - 1]], dtype = "float32")
             
    M = cv2.getPerspectiveTransform(rect, dst) 
    warped = cv2.warpPerspective(img, M, (maxWidth, maxHeight))      
     
    return warped
```

### 3.2. 함수 sq_detect()
- 영상내의 네개의 직선(contour)으로 둘러싸인 도형을 검출한다.
|Input|Output|
|:---:|   :---   |
|image|contour_image : contour을 그린 결과</br>warped_iamge : 투시변환 결과| 

```python
def sq_detect(image):
    origin = image.copy()
    gray = cv2.cvtColor(origin, cv2.COLOR_BGR2GRAY) # gray scale로 변환
    edge = cv2.Canny(gray, 75, 200) # Canny edge 적용하여 edge를 검출함
    kernel = np.ones((10,10), np.uint8)
    dilate = cv2.dilate(edge, kernel, iterations=1) # 없으면 미로검출 안됨
    # contour를 검출
    contours, hierarchy = cv2.findContours(dilate.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    contours = sorted(contours, key = cv2.contourArea, reverse = True)[:5]
    # 미로의 입구와 출구로 인한 왜곡과 A4용지의 구부러짐또는 빛에 의한 정보손실을 고려하여 approxPolyDP 사용
    for contour in contours:
        perimeter = cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, 0.02 * perimeter, True)
        if len(approx) == 4: # 4개의 윤곽을 찾았다면 종료
            document = approx
            break
    # 윤곽선의 정보를 기준으로
    warped_image = transform(origin, document.reshape(4, 2))
    # 기존의 이미지에 검출한 윤곽을 그림
    contour_image = cv2.drawContours(origin, [document], -1, (0, 255, 0), 2) 
    
    return contour_image, warped_image  # contour한 결과를 출력
```

```python
a4_contour, a4_warp = sq_detect(img)
```

## 5. 출발, 도착점 검출

- 필요한 함수
    - find_widest_region() : 뒤에서 설명
    - find_point() : 미로에 찍힌 출발점(초록색 점), 도착점(빨간색 점)의 위치를 검출한다.
    

### 5.1 함수 find_widest_region()
    - 함수 find_point()에서 빨간점과 도착점의 정확한 위치가 필요하다.
    - 노이즈가 포함된 어떤 matrix내에서 영역이 가장 큰 곳이 빨간점, 초록점이고 이 영역의 좌표값들을 얻기 위해 영상내에서 넓이가 가장 큰 영역의 좌표값들만 return하는 함수가 필요하다.

```python
def find_widest_region(mat):
    '''
    이진 영상에서 가장 큰 영역에 속하는 좌표값들을 반환
    이때 배경은 0 영역은 1
    '''
    map = mat.copy()
    map = map.astype(np.uint8)
    
    h, w = mat.shape
    dx = [-1, 1, 0, 0]
    dy = [0, 0, 1, -1]
    region = 2
    
    for x in range(h):
        for y in range(w):
            if map[x][y] == 1:
                
                q = deque([(x, y)])
                map[x][y] = region
                while q:
                    cx, cy = q.popleft()
                    map[cx][cy] = region
                    for k in range(4):
                        nx = cx + dx[k]
                        ny = cy + dy[k] 
                        if nx < 0 or ny < 0 or nx >=h or ny >=w:
                            continue
                        if map[nx][ny] == 0 or map[nx][ny] == region:
                            continue
                        if map[nx][ny] == 1:
                            q.append((nx, ny))
                            map[nx][ny] = region
                region += 1
    points = []
    size_of_region = []
    for r in range(2,region):
        p =np.where(map==r)
        points.append(p)
        size_of_region.append(len(p[0]))
    return points[np.argmax(size_of_region)]
```

### 5.2. 함수 find_point()
- 빨간색점과 초록색점은 조명의 방향의 영향을 많이 받아 RGB값만을 사용하여 정확한 한곳의 위치를 잡기는 어려웠다.
- 이에 HSV값을 활용하기로 하였고 밝기값(V)을 제외한 H, S 각각에 임계값을 정하여 노이즈를 포함하더라도 초록색과 빨간색 점의 위치를 잘 표현하도록하였다.
- 출발지점의 X, Y , 도착지점의 X, Y 좌표와 도착점, 출발점이 일부 지워진 이미지를 출력한다.

```python
def find_point(image):
    '''
    이미지에서 출발점 (초록점), 도착점 (빨간점)의 위치를 출력
    BGR 이미지를 입력으로 받고 HSV로 변환
    H값(0-180)을 사용하여 초록과 빨간 점의 위치 출력
    '''
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    rmap_h = hsv[:,:,0]<10  # H값이 0에 가까운 필셀들이 빨간색을 띤다.
    rmap_s = hsv[:,:,1]>120 # 채도가 120 이상되는 점들을 사용하여 노이즈를 줄인다.

    gmap_h = (hsv[:,:,0]>40) * (hsv[:,:,0]<80) # H값이 40과 80 사이인 픽셀들이 초록색을 띤다.
    gmap_s = hsv[:,:,1]>100 # 채도가 120 이상되는 점들을 사용하여 노이즈를 줄인다.
    
    # h, s에 대하여 두 조건을 만족하는 맵을 만들고 영역이 가장 큰 곳의 중앙 좌표를 출발점과 도착점으로 사용한다.
    # find_widest_region 함수에서 자세히 설명
    start = find_widest_region(rmap_h*rmap_s)
    end = find_widest_region(gmap_h*gmap_s)

    startX = np.median(start[0])
    startY = np.median(start[1])
    endX = np.median(end[0])
    endY = np.median(end[1])

    # 추가로 미로에서 출발 도착 지점에 해당하는 부분은 통로로 만들어줘야하기 때문에
    # end와 start에 담긴 좌표를 사용하여 image의 출발 도착 지점을 (255, 255, 255)로 만들어준다.
    img_erased = image.copy()
    cv2.circle(img_erased, (int(startY),int(startX)), 2, (255,255,255), thickness = 10)
    cv2.circle(img_erased, (int(endY),int(endX)), 2, (255,255,255), thickness = 10)
    
    plt.figure(figsize=(15, 15))
    plt.subplot(3, 3, 1); plt.title('rmap_h');plt.imshow(rmap_h, cmap='gray')
    plt.subplot(3, 3, 2); plt.title('rmap_s');plt.imshow(rmap_s, cmap='gray')
    plt.subplot(3, 3, 3); plt.title('rmap_hs');plt.imshow(rmap_h*rmap_s, cmap='gray')
    plt.subplot(3, 3, 4); plt.title('gmap_h');plt.imshow(gmap_h, cmap='gray')
    plt.subplot(3, 3, 5); plt.title('gmap_s');plt.imshow(gmap_s, cmap='gray')
    plt.subplot(3, 3, 6); plt.title('gmap_hs');plt.imshow(gmap_h*gmap_s, cmap='gray')
    plt.subplot(3, 3, 7); plt.title('img_erased');plt.imshow(img_erased, cmap='gray')
        
    return (int(startX), int(startY)), (int(endX), int(endY)), img_erased
```

```python
start, end, maze_erased = find_point(maze_warp)
```

> 결과를 보면 H값으로는 초록, 빨강의 색을 찾고, S값으로 검정색, 하얀색의 채도가 매우 작은 통로와 벽을 제외한 점의 위치를 알 수 있다.

## 6. 미로길 찾기
- 필요한 함수
    - binary_inv() : gray scale의 이미지를 이진화하여 벽과 통로를 구분한다.
    - block_outside() : 미로가 길을 바깥쪽으로 찾지 않도록 바깥의 영역을 막아준다.
    - maze_solver() : BFS 알고리즘을 사용하여 미로의 길을 찾고 정답지를 만든다.
    - mkgif() : gif파일로 만들어 저장한다.

### 6.1. 함수 binary_inv()
[참고링크](https://runebook.dev/ko/docs/scikit_image/api/skimage.filters#skimage.filters.threshold_local)

```python
def binary_inv(maze_erased):
    maze_gray = cv2.cvtColor(maze_erased, cv2.COLOR_BGR2GRAY)
    T = threshold_local(maze_gray, 11, offset = 10, method = "gaussian")
    maze_bin = (maze_gray<= T).astype("uint8")*255
    plt.figure(figsize=(15, 15))
    plt.imshow(maze_bin, cmap='gray')
    plt.show()
    return maze_bin
```

```python
maze_bin = binary_inv(maze_erased)
```

### 6.2. 함수 block_outside()
- 길을 밖으로 찾지 않도록 최외각선 밖의 영역을 막아줘야됨
- 네 꼭짓점에서 중앙을 향해서 검사하면서 벽(255)를 만날때까지 벽을 만들어준다.

```python
def block_outside(image):
    '''
    길을 밖으로 찾지 않도록 최외각선 밖의 영역을 막아줘야됨
    '''
    img = image.copy()
    h, w = image.shape
    lu = [0,0]
    ld = [h-1,0]
    ru = [0,w-1]
    rd = [h-1,w-1]
    while img[lu[0],lu[1]] == 0:
        img[lu[0], lu[1]] = 255
        img[lu[0], lu[1]+1] = 255
        img[lu[0]+1, lu[1]] = 255
        lu[0] += 1
        lu[1] += 1
    while img[ld[0],ld[1]] == 0:
        img[ld[0], ld[1]] = 255
        img[ld[0], ld[1]+1] = 255
        img[ld[0]-1, ld[1]] = 255
        ld[0] -= 1
        ld[1] += 1
    while img[ru[0],ru[1]] == 0:
        img[ru[0], ru[1]] = 255
        img[ru[0], ru[1]-1] = 255
        img[ru[0]+1, ru[1]] = 255
        ru[0] += 1
        ru[1] -= 1
    while img[rd[0],rd[1]] == 0:
        img[rd[0], rd[1]] = 255
        img[rd[0], rd[1]-1] = 255
        img[rd[0]-1, rd[1]] = 255
        rd[0] -= 1
        rd[1] -= 1
    return img
```

```python
maze_in = block_outside(maze_bin)
plt.figure(figsize=(15, 15))
plt.imshow(maze_in, cmap='gray')
plt.show()
```

### 6.3. 함수 mkgif()
- 현재 위치에 gif파일을 저장

```python
import os
import imageio
from PIL import Image
```

```python
def mkgif():
    path = [f"./gifs/{i}" for i in os.listdir("./gifs")]
    path.sort()
    paths = [ Image.open(i) for i in path]
    print(len(paths))
    imageio.mimsave('./result.gif', paths, fps=20)
```

### 6.4. 함수 maze_solver()
- BFS를 사용하여 미로의 길을 찾고 이진화 전의 미로 영상에 정답 경로를 표시한다.
- 추가로 gif파일로 만들어 저장한다.


```python
def maze_solver(maze, start_point, end_point, maze_warp):
    map = np.zeros_like(maze, np.float32)
    startx, starty = start_point
    endx, endy = end_point
    h, w = maze.shape
    labeled = maze_warp.copy()
    # 이동방향
    dx = [-1, 0, 1, 0]
    dy = [0, -1, 0, 1]

    q = deque([(startx, starty)])
    branch = []
    
    cnt = 0
    while q:
        x, y = q.popleft()
#         print(x, y)
        for i in range(4):
            nx = x+dx[i]
            ny = y+dy[i]
            if nx < 0 or nx >= h or ny < 0 or ny >=w:
                continue
            if nx == endx and ny == endy:
                print('finish!')
                branch.reverse()
                temx, temy, _, _ = branch[0]
                road = []
                for ix, iy, jx, jy in branch[1:]:
                    if temx == jx and temy == jy and map[temx][temy] - map[ix][iy] >0:
                        temx, temy = ix, iy
                        road.append((iy, ix))
                # draw
                road.reverse()
                for ix, iy in road:
                    cv2.circle(labeled, (ix, iy), 2, (0, 0, 255), thickness = 1)          
                    cnt += 1
                    if cnt%30 == 0:
                        cv2.imwrite(f'./gifs/img{cnt}.png', labeled)
#                 cv2.imshow('res', labeled)
#                 cv2.waitKey(0)
#                 cv2.destroyAllWindows()
                plt.figure(figsize=(15, 15))
                plt.imshow(labeled[:,:,::-1], cmap='gray')
                plt.show()
                mkgif()
                return
            if maze[nx][ny] != 0:
                continue
            elif maze[nx][ny] == 0:
                q.append((nx,ny))
                maze[nx][ny] = 255
                map[nx][ny] = map[x][y] + 1
                branch.append((x, y, nx, ny))
```

```python
maze_solver(maze_in, start, end, maze_warp)
```
